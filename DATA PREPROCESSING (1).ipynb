# Feature engineering

#### There are two particular columns ['Wage','Value'] are actually Continuous features, but these are being considered as Categorical features

```python
FIFADF['Wage'].head()
```

We can see that The 'Wage','Value' features dtype is object, the values have suffixes which are to be removed.

```python
FIFADF['Value'].head()
```

#### We can see that these are not correct values.
#### We have to perform some cleaning tasks to remove the suffixes to the values.

```python
def cleaning_task(df_value):
    try:
        value = float(df_value[1:-1])
        suffix = df_value[-1:]

        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value

FIFADF['Value'] = FIFADF['Value'].apply(cleaning_task)
FIFADF['Wage'] = FIFADF['Wage'].apply(cleaning_task)
```

#### Values are present in the form of strings (88+2), which depicts the performance of a player if he is played in the particular position

```python
def skillCorrection(val):
    if type(val) == str:
        s1 = val[0:2]
        s2 = val[-1]
        val = int(s1) + int(s2)
        return val
    
    else:
        return val
```

```python
skill_columns = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
                      
for col in skill_columns:
    FIFADF[col] = FIFADF[col].apply(skillCorrection)
```

### Detecting Outliers in Continuous Features

```python
DATA_QUALITY_R
```

#### We will remove outlier with using the below function, We will send FIFADF and continuous features dataset as input. 

```python
def outlier_removal(df_in, col_name):
    for i in col_name.columns:
        q1 = df_in[i].quantile(0.25)
        q3 = df_in[i].quantile(0.75)
        iqr = q3-q1 #Interquartile range
        fence_low  = q1-1.5*iqr
        fence_high = q3+1.5*iqr
        df_out = df_in.loc[(df_in[i] > fence_low) & (df_in[i] < fence_high)]
    return df_out
```

```python
FIFADF_Pure=outlier_removal(FIFADF,Continuous_features)
```
